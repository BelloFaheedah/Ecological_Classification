<h2>Classification Using Ecological Dataset</h2>


<h5>Importing the dataset</h5>

```python
import pandas as pd
dataset=pd.read_csv('ecology_dataset.csv')
dataset
```

<p>Checking out for missing values and duplicated rows and dropping the unwanted column</p>

```python
dataset.drop(['stab'], axis=1)
```

```python
dataset.isnull().sum()
```

<p>From the results shown above, it is apparent that the dataset is devoid of missing values and duplicated values</p>

<p>Data Preprocessing</p>

```python
x=dataset.iloc[:,0:12].values
y=dataset.iloc[:,13:14].values
```

```python
#splitting the dataset into the ration 80:20
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
#transforming the dataset using standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(x_train,y_train)
scaler.transform(x_test)
```

<h4>Random Forest Classifier</h4>

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
# define the model
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
# evaluating the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

<p>This explains that a classification accuracy of about 84.7% has been achieved using the random forest classifier</p>

<h4>Extra Trees Classifier</h4>

```python
# evaluate extra trees algorithm for classification
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import ExtraTreesClassifier
# define the model
xmodel = ExtraTreesClassifier()
xmodel.fit(x_train,y_train)
```

```python
# evaluating the model using repeated stratified K_fold
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(xmodel, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.4f (%.4f)' % (mean(n_scores), std(n_scores)))
```

<p>Just like Random Forest, the accuracy observed when using extra classifiers is 84.1% </p>

<h4>XGBoost</h4>

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
#fitting the XGboost to the training set
xgmodel = XGBClassifier()
xgmodel.fit(x_train, y_train)
```

```python
#Evaluating the model
y_pred = xgmodel.predict(x_test)
# evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

<p>The observed accuracy from the XGBoost model is 91.95%, and this is better.</p>

<h4>Randomized Search Cross validation</h4>

```python
#Improving the extra classifiers using hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
rsmodel=ExtraTreesClassifier()
n_estimators= [50, 100, 300, 500, 1000]
min_samples_split = [2, 3, 5, 7, 9]
min_samples_leaf = [1, 2, 4, 6, 8]
max_features = ['auto', 'sqrt', 'log2', None] 
hyperparameter_grid = {'n_estimators': n_estimators,
                       'min_samples_leaf': min_samples_leaf,
                       'min_samples_split': min_samples_split,
                       'max_features': max_features}
rscv=RandomizedSearchCV(rsmodel,hyperparameter_grid)
```

```python
best_model=rscv.fit(x_train,y_train)
best_model.best_estimator_
```

<p>The model above will produce the best precision since it has been iterated severally on different subsets of data </p>
